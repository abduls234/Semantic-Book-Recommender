```python
import os
import numpy as np
from sentence_transformers import SentenceTransformer
from langchain.document_loaders import TextLoader
from langchain.text_splitter import CharacterTextSplitter
import faiss
import pickle
```

```python
# Load your text data
raw_documents = TextLoader("tagged_description.txt", encoding="utf-8").load()
text_splitter = CharacterTextSplitter(chunk_size=0, chunk_overlap=0, separator="\n")
documents = text_splitter.split_documents(raw_documents)

texts, isbns = [], []
for doc in documents:
    line = doc.page_content
    if ":" in line:
        isbn, desc = line.split(":", 1)
        texts.append(desc.strip())
        isbns.append(isbn.strip())
```

```python
model = SentenceTransformer("all-MiniLM-L6-v2")

def batch_embed(texts, batch_size=100):
    embeddings = []
    for i in range(0, len(texts), batch_size):
        batch = texts[i:i+batch_size]
        emb = model.encode(batch, show_progress_bar=True)
        embeddings.extend(emb)
    return np.array(embeddings)

print("🔄 Embedding started...")
embeddings = batch_embed(texts)
print("✅ Embedding done.")
```

```python
# Define Drive paths
drive_path = "/content/drive/MyDrive/book_vector_db"
os.makedirs(drive_path, exist_ok=True)

# Create and save FAISS index
dimension = embeddings.shape[1]
index = faiss.IndexFlatL2(dimension)
index.add(embeddings)

# Save FAISS index
faiss.write_index(index, os.path.join(drive_path, "book_index.faiss"))

# Save metadata (for reverse lookup)
with open(os.path.join(drive_path, "isbns.pkl"), "wb") as f:
    pickle.dump(isbns, f)

with open(os.path.join(drive_path, "texts.pkl"), "wb") as f:
    pickle.dump(texts, f)

print("✅ Vector DB saved to Google Drive.")
```

```python
# Reload index and metadata
index = faiss.read_index(os.path.join(drive_path, "book_index.faiss"))

with open(os.path.join(drive_path, "isbns.pkl"), "rb") as f:
    isbns = pickle.load(f)

with open(os.path.join(drive_path, "texts.pkl"), "rb") as f:
    texts = pickle.load(f)

# Search example
query = "mental health awareness"
query_embedding = model.encode([query])
D, I = index.search(np.array(query_embedding), k=5)

# Show results
for idx in I[0]:
    print(f"📖 ISBN: {isbns[idx]}")
    print(f"📝 Desc: {texts[idx]}")
    print("---")
```

```python
import faiss
import pickle
from sentence_transformers import SentenceTransformer
import numpy as np

# 1. Load FAISS index
index = faiss.read_index("book_vector_db/book_index.faiss")

# 2. Load metadata
with open("book_vector_db/texts.pkl", "rb") as f:
    texts = pickle.load(f)

with open("book_vector_db/isbns.pkl", "rb") as f:
    isbns = pickle.load(f)


```

```python
import pandas as pd
books_df = pd.read_csv("books_cleaned.csv")

# Convert your FAISS-side isbns (strings) to integers safely
cleaned_isbns = []
for isbn in isbns:
    try:
        cleaned_isbns.append(int(isbn))
    except ValueError:
        cleaned_isbns.append(None)  # or skip if preferred

# Now use cleaned_isbns instead of isbns directly
model = SentenceTransformer("all-MiniLM-L6-v2")

def search_books(query, k=2):
    query_vec = model.encode([query]).astype("float32")
    D, I = index.search(query_vec, k)

    matched_isbns = [cleaned_isbns[idx] for idx in I[0] if cleaned_isbns[idx] is not None]
    matched_rows = books_df[books_df['isbn13'].isin(matched_isbns)]

    return matched_rows
```

```python
search_books("a book about world war")
```
